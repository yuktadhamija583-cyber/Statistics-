#Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?

-> A confidence interval is a range of values that is likely to contain the true population parameter (like mean or proportion) with a certain level of confidence (e.g., 95%).

In simple words:

A confidence interval tells us how sure we are about our estimate.

Example (95% CI):
[12.4 , 18.6]
This means we are 95% confident the true population mean lies between 12.4 and 18.6.

> What is Margin of Error (ME)?

The margin of error is the amount added and subtracted from the sample estimate to create the confidence interval.

Formula (general idea):

Confidence Interval = Sample Estimate Â± Margin of Error


ðŸ”¹ How Margin of Error Influences CI Width and Interpretation

**1ï¸âƒ£ Margin of Error increases â†’ CI becomes wider **bold text**

A larger margin of error gives a broader range.

This means your estimate is less precise.

But it increases your confidence that the true value is inside the interval.

Example:
CI = 50 Â± 10 â†’ [40, 60] (wide, less precise)

**2ï¸âƒ£ Margin of Error decreases â†’ CI becomes narrower**

A smaller margin of error gives a tighter range.

This means your estimate is more precise.

But too narrow a CI may reduce actual confidence if the margin is too small.

Example:
CI = 50 Â± 2 â†’ [48, 52] (narrow, more precise)

**Factors that affect margin of error : **

Margin of error depends on:

Sample size (n)

Higher variability â†’ larger ME â†’ wider CI

Confidence level (90%, 95%, 99%)

**Example:**

âœ… How ANOVA Extends Hypothesis Testing to More Than Two Groups

âœ… Assumptions of ANOVA

#Question 6: Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset.

->

```python
import numpy as np
from scipy.stats import norm

# ----- Sample Data -----
data = [52, 55, 49, 50, 53, 54, 56, 51]   # example sample
population_mean = 50                      # H0: Î¼ = 50
population_std = 5                        # known population standard deviation
alpha = 0.05                              # significance level

# ----- Step 1: Calculate sample statistics -----
sample_mean = np.mean(data)
n = len(data)

# ----- Step 2: Compute Z-statistic -----
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# ----- Step 3: Compute P-value (two-tailed test) -----
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# ----- Step 4: Print Results -----
print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("P-value:", p_value)

# ----- Step 5: Interpretation -----
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

Interpretation (How this works)
Null Hypothesis (Hâ‚€):

The population mean is equal to the claimed mean.
Hâ‚€: Î¼ = 50

Alternative Hypothesis (Hâ‚):

The population mean is different from the claimed mean.
Hâ‚: Î¼ â‰  50 (two-tailed)

What the program outputs:

Z-statistic â†’ how many standard errors the sample mean is away from the population mean

P-value â†’ probability of getting such an extreme sample if Hâ‚€ were true

Interpretation

If p-value < Î± â†’ Reject Hâ‚€ (significant difference)

If p-value â‰¥ Î± â†’ Fail to reject Hâ‚€ (no significant difference)

#Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_confidence_interval(data, confidence=0.95):
    # Calculate mean
    mean = np.mean(data)

    # Standard error of the mean
    sem = stats.sem(data)

    # t-critical value
    t_crit = stats.t.ppf((1 + confidence) / 2, df=len(data) - 1)

    # Margin of error
    margin = t_crit * sem

    # Confidence interval bounds
    lower = mean - margin
    upper = mean + margin

    # Visualization
    plt.figure(figsize=(8,5))
    plt.errorbar(
        x=[0],
        y=[mean],
        yerr=[margin],
        fmt='o',
        capsize=10
    )
    plt.title(f"{int(confidence*100)}% Confidence Interval for Sample Mean")
    plt.ylabel("Value")
    plt.xticks([])
    plt.show()

    return mean, lower, upper


```

#Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.

```python
import numpy as np
from scipy.stats import chisquare

# Observed frequencies
observed = np.array([30, 20, 25, 25])

# Expected frequencies (equal distribution example)
expected = np.array([25, 25, 25, 25])

# Perform Chi-square goodness-of-fit test
chi_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print("Chi-square Statistic:", chi_stat)
print("p-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis â€” observed distribution differs significantly from expected.")
else:
    print("Result: Fail to reject the null hypothesis â€” no significant difference between observed and expected.")
```
